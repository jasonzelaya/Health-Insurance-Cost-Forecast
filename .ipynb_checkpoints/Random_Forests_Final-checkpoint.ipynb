```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Import unmodified csv file
insurance = pd.read_csv("insurance.csv")
```

```python
# Create a heatmap to visualize any correlations between variables from the original, unmodifiied dataset
plt.figure(figsize=(15, 15))
sns.heatmap(insurance.corr(), annot=True)
```

```python
# Import csv file
df = pd.read_csv("insurance_cleaned.csv")

# Replace the sex_female column name with sex as a column name and smoker_no columns name to smoker as the column name for 
# consistency 
# 1 means the beneficiary is female and 0 means the beneficiary is male
df.rename(columns={'sex_female':'sex'}, inplace=True)
# 1 means the beneficiary does not smoke and 0 means the beneficiary does smoke
df.rename(columns={'smoker_no':'smoker'}, inplace=True)

# Display first five rows of the dataframe
df.head()
```

```python
# Drop the Unnamed: 0 column that appeared during csv import
df = df.drop(columns='Unnamed: 0')

# Display first five rows of the dataframe
df.head()
```

```python
# assign the target variable to the values in the charges column
target = df.charges
target
```

```python
# Remove the columns that will not be used as independent values
df_data = df.drop(['charges', 'sex_male', 'smoker_yes'], axis = 1)

# Display first five rows of the dataframe
df_data.head()
```

```python
# Print the maximum value in the charges column
print(df['charges'].max())

# Print statistic values for all columns in the df dataframe
print(df.describe())

# Print a concise summary of the df dataframe
print(df.info())
```

```python
# Check the correlation between data
df.corr()
```

```python
# Create a heatmap to visualize any correlations between variables from the cleaned dataset
plt.figure(figsize=(15, 15))
sns.heatmap(df.corr(), annot=True)
```

```python
data_x = df.drop(['charges', 'sex_male', 'smoker_yes'], axis = 1)
data_x = list(data_x.columns.values)
data_x
```

```python
# Convert to numpy array
data = df_data
data = np.array(data)
data
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
print('Training Data Shape:', X_train.shape)
print('Training Target Shape:', y_train.shape)
print('Testing Data Shape:', X_test.shape)
print('Testing Target Shape:', y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Instantiate model with 100 decision trees
rf = RandomForestRegressor(n_estimators = 100, random_state = 42)
rf
```

```python
# Train the model on training data
rf.fit(X_train, y_train);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(X_test)
predictions
```

```python
# Calculate the absolute errors
errors = abs(predictions - y_test)
errors
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
mape
```

```python
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
# Get numerical feature importances
importances = list(rf.feature_importances_)
importances
```

```python
# List of tuples with variable and importance
data_importances = [(data, round(importance, 2)) for data, importance in zip(data_x, importances)]
data_importances
```

```python
# Sort the feature importances by most important first
data_importances = sorted(data_importances, key = lambda x: x[1], reverse = True)
data_importances
```

```python
# Print out the feature and importances 
[print('Variable: {:20} Importance: {}'.format(*importance)) for importance in data_importances]
```

```python
# Set the style
plt.style.use('fivethirtyeight')

# list of x locations for plotting
x_values = list(range(len(importances)))

# Make a bar chart
plt.bar(x_values, importances, orientation = 'vertical')

# Tick labels for x axis
plt.xticks(x_values, data_x, rotation='vertical')

# Axis labels and title
plt.ylabel('Importance'); plt.xlabel('Variable'); plt.title('Variable Importances');
```

```python
# New random forest with only the two most important variables
rf_most_important = RandomForestRegressor(n_estimators= 100, random_state=42)
rf_most_important
```

```python
# Extract the two most important features
important_indices = [data_x.index('smoker'), data_x.index('bmi'), data_x.index('age'), 
                     data_x.index('children'), data_x.index('sex'), data_x.index('region_northeast')]
train_important = X_train[:, important_indices]
test_important = X_test[:, important_indices]

print(important_indices)
print(train_important)
print(test_important)
```

```python
# Train the random forest
rf_most_important.fit(train_important, y_train)
```

```python
# Make predictions and determine the error
predictions_important = rf_most_important.predict(test_important)
predictions_important
```

```python
errors_important = abs(predictions_important - y_test)
errors_important
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, predictions_important))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, predictions_important))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(y_test, predictions_important)))
```

```python
# Calculate mean absolute percentage error (MAPE)
mapes = 100 * (errors_important / y_test)
mapes
```

```python
mapes_important = np.mean(100 * (errors_important / y_test))
accuracy_important = 100 - mapes

print('Accuracy:', accuracy_important, '%.')
```

```python
# Plot the Residuals for the Training and Testing data with a histogram side bar
from sklearn.linear_model import Ridge
from yellowbrick.regressor import ResidualsPlot

# Instantiate the linear model and visualizer
ridge = Ridge()
visualizer = ResidualsPlot(ridge)

visualizer.fit(X_train, y_train)  # Fit the training data to the model
visualizer.score(X_test, y_test)  # Evaluate the model on the test data
visualizer.poof()                 # Draw/show/poof the data
```
